2.2 Dataset Information

This dataset focuses on health indicators related to heart disease and contains the following features:

* HeartDiseaseorAttack: Indicates whether the individual has heart disease or has experienced a heart attack (1 for yes, 0 for no).
* HighBP: Denotes whether the individual has high blood pressure (1 for yes, 0 for no). High blood pressure is a significant risk factor for heart disease.
* HighChol: Indicates if the individual has high cholesterol levels (1 for yes, 0 for no). High cholesterol can lead to plaque buildup in arteries, increasing the risk of heart disease.
* CholCheck: Reflects whether the individual has had their cholesterol checked (1 for yes, 0 for no). Regular checks can help manage cholesterol levels effectively.
* BMI: Body Mass Index, a measure of body fat based on height and weight. It can indicate obesity, which is a risk factor for heart disease.
* Smoker: Indicates whether the individual is a smoker (1 for yes, 0 for no). Smoking is a well-known risk factor for various heart-related issues.
* Stroke: Reflects whether the individual has had a stroke (1 for yes, 0 for no). A history of stroke can indicate higher cardiovascular risk.
* Diabetes: Indicates whether the individual has diabetes (1 for yes, 0 for no). Diabetes significantly increases the risk of heart disease.
* PhysActivity: Denotes whether the individual engages in physical activity (1 for yes, 0 for no). Regular physical activity is essential for heart health.
* Fruits: Indicates whether the individual consumes fruits regularly (1 for yes, 0 for no). A diet rich in fruits can lower heart disease risk.
* Veggies: Reflects whether the individual consumes vegetables regularly (1 for yes, 0 for no). Like fruits, vegetables are vital for cardiovascular health.
* HvyAlcoholConsump: Indicates whether the individual engages in heavy alcohol consumption (1 for yes, 0 for no). Excessive alcohol can lead to heart issues.
* AnyHealthcare: Reflects whether the individual has access to healthcare services (1 for yes, 0 for no). Access to healthcare is crucial for regular check-ups and preventive care.
* NoDocbcCost: Indicates whether cost prevents the individual from seeing a doctor (1 for yes, 0 for no). Financial barriers can impact health outcomes.
* GenHlth: General health rating (on a scale, often from 1 to 5). This self-reported measure can provide insight into overall health.
* MentHlth: Reflects mental health status (on a scale, often from 1 to 30). Mental health can influence physical health, including heart disease risk.
* PhysHlth: Indicates physical health status (on a scale, often from 1 to 30). Similar to mental health, physical health status can impact overall cardiovascular risk.
* DiffWalk: Indicates whether the individual has difficulty walking (1 for yes, 0 for no). Mobility issues can be linked to overall health and cardiovascular risk.
* Sex: Indicates the individual's gender (usually 1 for male, 0 for female). Gender can play a role in heart disease prevalence and risk factors.
* Age: Age of the individual, often a critical factor in heart disease risk, as older age is associated with increased risk.
* Education: Level of education attained. Education can correlate with health literacy and access to healthcare.
* Income: Indicates the income level of the individual. Socioeconomic status can influence health outcomes, including access to healthcare and healthy lifestyle choices.

```python
bmi_bins = [0, 18.5, 24.9, 29.9, float('inf')]
bmi_labels = ['Underweight', 'Normal', 'Overweight', 'Obese']
df['BMI'] = pd.cut(df['BMI'], bins= bmi_bins, labels = bmi_labels, right=False)
                        
mental_health_bins = [0, 7, 14, 21, 30.1]  # Mental health score ranges
mental_health_labels = ['Low', 'Mild', 'Moderate', 'Severe']  # Labels for the bins

            # Apply pd.cut to categorize the 'Mental_Health_Score' column
df['Mental_Health_Category'] = pd.cut(df['MentHlth'], 
                                    bins=mental_health_bins, 
                                    labels=mental_health_labels, 
                                    right=False)
                        
physical_health_bins = [0, 7, 14, 21, 30.1]  # Mental health score ranges
physical_health_labels = ['Poor', 'Fair', 'Good', 'Excellent']  # Labels for the bins

# Apply pd.cut to categorize the 'Mental_Health_Score' column
df['Physical_Health_Category'] = pd.cut(df['PhysHlth'], 
                                    bins=physical_health_bins, 
                                    labels=physical_health_labels, 
                                    right=False)


new_df = df.drop(['CholCheck','NoDocbcCost','Age','MentHlth','PhysHlth','Education','Income'], axis=1, inplace=True)
```
